```python
#bar char with most spend model
price_model = cell.nlargest(10, ['price']).set_index('model')['price']
sns.barplot(x=price_model.index, y=price_model)
plt.title('Top 10 most spend model')
plt.xlabel('model')
plt.ylabel('price')
plt.xticks(rotation=90)
```

```python
#ram by model
ram_model = cell.nlargest(10, ['RAM']).set_index('model')['RAM']
sns.barplot(x=ram_model.index, y=ram_model)
plt.title('Top 10 best ram by model')
plt.xlabel('model')
plt.ylabel('RAM')
plt.xticks(rotation=90)
```
